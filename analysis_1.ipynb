```python
Q1 = ds['Sales'].quantile(0.25)
Q3 = ds['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower_lim = Q1 - 1.5 * IQR
upper_lim = Q3 + 1.5 * IQR
outliers_sales = ds[(ds['Sales'] < lower_lim) | (ds['Sales'] > upper_lim)]
len(outliers_sales)
```

```python
Q1 = ds['Profit'].quantile(0.25)
Q3 = ds['Profit'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers_for_profit = ds[(ds['Profit'] < lower_limit) | (ds['Profit'] > upper_limit)]
len(outliers_for_profit)
```
