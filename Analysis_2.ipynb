```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\sathi\OneDrive\Desktop\superstore.csv", encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
df.head()
```

```python
df = df.dropna(subset=['Order Date', 'Ship Date'])
```

```python
df.isnull().sum()
```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0)
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce').fillna(0)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').fillna(0)
df['Discount']=pd.to_numeric(df['Discount'],errors='coerce').fillna(0)
```

```python
categorical_cols = ['Ship Mode', 'Customer Name', 'Segment', 'Region', 'City', 'Country','State','Category','Sub-Category','Product Name']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
df.head()
```

```python
df['Category'] = df['Category'].str.strip().str.title()
df['Sub_Category'] = df['Sub-Category'].str.strip().str.title()
df['Segment'] = df['Segment'].str.strip().str.title()
df['Region'] = df['Region'].str.strip().str.title()
df['Ship_Mode'] = df['Ship Mode'].str.strip().str.title()
df['Postal_Code'] = df['Postal Code'].astype(str).str.strip()
df['Product_Name'] = df['Product Name'].str.strip().str.title()


```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')


```

```python
df.to_csv("cleaned_superstore.csv", index=False)
df.head()
```

```python

import pandas as pd



col_names = df.columns.tolist()
col_types = [str(df[col].dtype) for col in df.columns]

# ----- Description master list -----
description_master = [
    "Unique Row ID", "Order Identifier", "Order date", "Shipping date",
    "Shipping mode", "Customer ID", "Customer Name", "Customer Segment",
    "Country", "City", "State", "Postal Code", "Region",
    "Product ID", "Product Category", "Product Sub-Category",
    "Product Name", "Sales amount", "Quantity ordered", "Discount applied",
    "Profit amount"
]

# ----- Cleaning Steps master list -----
cleaning_master = [
    "Original ID", "No changes", "Converted to datetime", "Converted to datetime",
    "Standardized values", "No changes", "No changes", "Filled missing, standardized case",
    "No changes", "No changes", "No changes", "No changes", "Standardized case",
    "No changes", "Standardized case", "Standardized case",
    "No changes", "Converted to float, filled missing", "Converted to int, filled missing",
    "Converted to float", "Converted to float, filled missing"
]

# ✨ Auto-match lengths (truncate or pad)
def adjust_list(master_list, target_len):
    if len(master_list) >= target_len:
        return master_list[:target_len]
    else:
        return master_list + ["No description available"] * (target_len - len(master_list))

desc_list = adjust_list(description_master, len(col_names))
clean_list = adjust_list(cleaning_master, len(col_names))

# Create dictionary dataframe
data_dict = pd.DataFrame({
    "Column Name": col_names,
    "Data Type": col_types,
    "Description": desc_list,
    "Cleaning Steps": clean_list
})

# Export
data_dict.to_excel("data_dictionary.xlsx", index=False)

print("Data dictionary saved as data_dictionary.xlsx")

```
